### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
input_path = "../WeatherPy/city_weather_data.csv"
weather_data = pd.read_csv(input_path)
weather_data
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
center = (37.105706,	58.508362)
fig = gmaps.figure(map_type='HYBRID', zoom_level=3, center=center)
heatmap_layer = gmaps.heatmap_layer(
    weather_data[['Lat', 'Lng']], weights=weather_data['Humidity'], dissipating=False,
    max_intensity=100, point_radius = 3
)
fig.add_layer(heatmap_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
best_weather = weather_data.loc[(weather_data['Max_Temp'] <= 22) &
(weather_data['Wind_Speed'] < 8) & (weather_data['Humidity'] < 40)]
best_weather.dropna(how='any')
best_weather
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = best_weather.copy()
hotel_df['Hotel Name'] = None
hotel_df['Hotel Lat'] = None
hotel_df['Hotel Lng'] = None
hotel_df.head()
```

```python
# Google Places API Query
# Set Parameters for search
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "radius":5000,
    "type": "lodging",
    "key": g_key
}

# Loop through Places API to find hotels
for index, row in hotel_df.iterrows():
    try:
        params['location'] = f"{row['Lat']}, {row['Lng']}"
        response = requests.get(base_url, params=params).json()
        hotels = response["results"][0]
        hotel_df.loc[index, "Hotel Name"] = hotels['name']
        hotel_df.loc[index, "Hotel Lat"] = hotels['geometry']['location']['lat']
        hotel_df.loc[index, "Hotel Lng"] = hotels['geometry']['location']['lng']
    except:
        print("No Hotels Nearby")
```

```python
# Drop locations with no nearby hotels
hotel_df = hotel_df.dropna(how='any')
hotel_df.head()
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{Name}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer
marker_coord = hotel_df[['Hotel Lat', 'Hotel Lng']]
markers = gmaps.marker_layer(marker_coord, info_box_content=hotel_info)
fig.add_layer(markers)
fig
```
